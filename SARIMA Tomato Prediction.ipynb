**Market 1**

```python
import pandas as pd
import numpy as np
df=pd.DataFrame
df=pd.read_excel('Market 1 Tomato.xlsx')
df.set_index(df.iloc[:,0],inplace=True)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# # Generate some sample demand data (replace with your own data)
# np.random.seed(42)
# demand_data = np.random.randint(50, 150, 100)  # Replace with your actual data

# # Create a date range for the time series (assuming monthly data)
# start_date = pd.to_datetime('2020-01-01')
# date_range = pd.date_range(start_date, periods=len(demand_data), freq='M')

# Create a pandas DataFrame with the demand data
demand_df = pd.DataFrame
demand_df=df
demand_df.set_index(demand_df.iloc[:,0], inplace=True)

# Split the data into training and testing sets
# train_size = int(len(demand_df) * 0.8)  # 80% for training
# train_data = demand_df[118:848]
train_data = demand_df[0:738]
# test_data = demand_df[1579:1827]
test_data = demand_df[738:]

# Define SARIMA model parameters (replace with your chosen values)
p = 2  # Autoregressive order #4 0 4 4 0 4
d = 1  # Differencing order
q = 2  # Moving average order
P = 2  # Seasonal autoregressive order
D = 2  # Seasonal differencing order
Q = 2  # Seasonal moving average order
s = 12  # Seasonal period (assuming monthly data with yearly seasonality)

# Fit the SARIMA model to the training data
model = SARIMAX(train_data['Arrival'], order=(p, d, q), seasonal_order=(P, D, Q, s))
model.initialize_approximate_diffuse()
results = model.fit()

# Generate forecasts for the test data
forecast_start = len(train_data)
forecast_end = len(train_data) + len(test_data) - 1
forecast = results.get_prediction(start=forecast_start, end=forecast_end, dynamic=False)

# Get the predicted values and confidence intervals
predicted_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Plot the observed and predicted values
plt.figure(figsize=(10, 6))
# plt.plot(train_data.index, train_data['Demand'], label='Training Data')
plt.plot(test_data.index, test_data['Arrival'], label='Test Data')
plt.plot(df.index[738:], predicted_values, color='red', label='Predicted Data')
plt.fill_between(df.index[738:], confidence_intervals['lower Arrival'], confidence_intervals['upper Arrival'], color='pink')
plt.title('SARIMA Demand Forecasting')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()

# Calculate and print evaluation metrics (e.g., RMSE, MAE)
from sklearn.metrics import mean_squared_error, mean_absolute_error,mean_absolute_percentage_error

rmse = np.sqrt(mean_squared_error(test_data['Arrival'], predicted_values))
mae = mean_absolute_error(test_data['Arrival'], predicted_values)
mape=mean_absolute_percentage_error(test_data['Arrival'], predicted_values)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAPE): {mape}")
```
